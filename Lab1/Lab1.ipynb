```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**2+5

x1 = np.arange(-6, 6, 0.01)
x2 = np.arange(-0, 5, 0.01)
x3 = np.arange(-1, 1, 0.01)
plt.plot(x1, f(x1), x2, f(x2), x3, f(x3))
plt.grid()
plt.title('Funkcja f na przedziałach')
plt.xlabel('Argumenty x')
plt.ylabel('Wartości y')
plt.legend(['(-1,1)', '(-6, 6)', '(0,5)'])
plt.show()
```
