Import niezbędnych bibliotek oraz danych z oryginalnego pliku billboard.csv

```python
import pandas as pd
import numpy as np
billboard = pd.read_csv('../Original Data/billboard.csv')
# billboard
```

Utworzenie nowej tabeli na przetworzone dane

```python
tab = pd.DataFrame({
    billboard.columns[1]: [billboard.values[0][1]],
    billboard.columns[2]: [billboard.values[0][2]],
    billboard.columns[3]: [billboard.values[0][3]],
    billboard.columns[4]: [billboard.values[0][4]],
    billboard.columns[5]: [billboard.values[0][5]],
    billboard.columns[6]: [billboard.values[0][6]],
    'week': int(1),
    'rank': int(billboard.values[0][7])
})
```

Przejście dwiema pętlami po wszytskich utowrach i wszystkich tygodniach, aby zapisać niezerowe dane do przetworzonej tabeli

```python
for i in range(billboard.shape[0]):
    for j in range(7, billboard.shape[1]):
        if not np.isnan(billboard.iloc[i][j]):
            if i != 0 or j != 7:
                tab.loc[tab.shape[0],:] = [
                    billboard.values[i][1],
                    billboard.values[i][2],
                    billboard.values[i][3],
                    billboard.values[i][4],
                    billboard.values[i][5],
                    billboard.values[i][6],
                    j-6,
                    billboard.iloc[i][j]
            ]
tab
```

Zapis otrzymanej przetworzonej tabeli do odpowiedniego folderu w dwóch wersjach rozszerzeń: .csv oraz .xlsx

```python
tab.to_csv("../Analysis Data/tab.csv")
tab.to_excel("../Analysis Data/tab.xlsx")
```
